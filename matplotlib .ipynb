```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# first data set
x = np.arange(0,100,10)
y = np.cos(x)
plt.plot(x,y)
plt.title("Cosine Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# second data set using random
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)
size = np.random.randint(30,50,30)

# graph creation
plt.scatter(x,y, c = colors , s = size)
plt.title("Point Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# third data set 
data = np.random.randn(5000)

# graph creation
plt.hist(data,bins= 30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4th data set 
x = ["1990","1995","2000","2005","2010"]
y = np.random.randint(1,10,5)
# graph creation
plt.bar(x,y)
plt.title("Column Chart")
plt.xlabel("Years")
plt.ylabel("Values")
plt.show()
```

```python
# 5th data set 
sizes = [30,25,20,15,10,5,0]
plt.pie(sizes)
plt.title("Pie Chart")
plt.show()
```
